```python
from random import random
import numpy as np
from matplotlib import pyplot
C1 = 0
C2 = random()*2*np.pi
beta = random()*10**(-3)
alpha = random()*beta**3.65
rad = np.arange(0,np.pi,0.01)
r0 = 1/beta*np.sin(rad+C1)+alpha*C2*np.cos(rad+C1)
r0 = r0+alpha/beta**4*(np.sin(3*rad+C1)+4*np.sin(rad+C1)*np.log(abs(np.cos(rad+C1))))
r0 = r0+alpha/beta**4*(np.sin(rad-C1)*(5-4*np.log(abs(np.cos(rad+C1)))))
r0 = r0-4*rad*np.cos(rad+C1)*(1+2*np.cos(2*C1))
pyplot.polar(rad,1/beta*np.sin(rad+C1))
pyplot.polar(rad,r0)
```
